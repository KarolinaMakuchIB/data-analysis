# Principal component analysis

In this small project I am using PCA to find principal components in coffee database. Database contains two kinds of coffee: arabica and robusta.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

dataset = pd.read_csv("kawa.txt", delim_whitespace = True, header = 0)
print(dataset.head())
X=dataset.values[:, 1:12]
y= dataset['Kawa']

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

pca = PCA()
PC = pca.fit_transform(X)

print('\nExplained variance ratio: ', pca.explained_variance_ratio_)
sns.barplot(x=np.arange(1,12), y= pca.explained_variance_ratio_, color='blue')
plt.xlabel('Explained variance')
plt.ylabel('Component number')
plt.show()

print('\nResults - projection of objects on the plane of the main component 1 and 2:')
sns.scatterplot(x=PC[:,0], y=PC[:,1], sizes=50, palette = 'viridis', hue=y)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
